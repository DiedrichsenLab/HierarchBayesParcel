```python
import os # to handle path information
import nibabel as nb
import numpy as np
import h5py
import pandas as pd
import surfAnalysisPy as surf
import matplotlib.pyplot as plt
```

```python
hem = ['L','R']
hem_name = ['CortexLeft','CortexRight']
```

```python
baseDir = '/Users/jdiedrichsen/Data/fs_LR_32'
flatsurf = []
inflsurf = [] 
glasser = []

for h,hemN in enumerate(hem):
    flatname = os.path.join(baseDir,'fs_LR.32k.' + hemN + '.flat.surf.gii')
    inflname = os.path.join(baseDir,'fs_LR.32k.' + hemN + '.inflated.surf.gii')
    labname = os.path.join(baseDir,'Glasser_2016.32k.' + hemN + '.label.gii')
    flatsurf.append(nb.load(flatname))
    inflsurf.append(nb.load(inflname))
    glasser.append(nb.load(labname))
```

```python
from nilearn import plotting
coords = inflsurf[0].agg_data('pointset')
faces = inflsurf[0].agg_data('triangle')
L = glasser[0].agg_data()
L 
```

```python
plotting.plot_surf([coords,faces],L,darkness=0.3,hemi='left')
```

```python
plotting.view_surf([coords,faces],L)
```
